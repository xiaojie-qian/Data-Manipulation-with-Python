## 4. Broadcasting
- Compatibility rules: Numpy compares sets of array dimensions from right to left 
- Two dimensions are compatible when ...
  - One of them has a length of one or 
  - thez are of equal lengths
- All dimension sets must be compatible
- (10, 5) and (10, 1), (10, 5) and (5, ) are compatible 
- (10, 5) and (5, 10) are not compatible

```python
# Broadcasting rows: 1D array copies the 1st row as the 2rd row 
array = np.arange(10).reshape((2, 5))
array + np.array([0, 1, 2, 3, 4])
```

```python
# two arrays are not compatible 
array = np.arange(10).reshape((2, 5))
array + np.array([0, 1])
```

```python
array = np.arange(10).reshape((2, 5))
array + np.array([0, 1]).reshape((2,1))
```

### 5.3 Stacking and splitting 

```python
rgb = np.array([[[255,0,0], [255,255,0], [255,255,255]],
               [[255,0,255], [0,255,0], [0,255,255]],
               [[0,0,0], [0,255,255], [0,0,255]]])
red_array = rgb[:, :, 0] # the first column of each array 
green_array = rgb[:, :, 1] # the second column of each array 
blue_array = rgb[:, :, 2]
print(rgb.shape)
print(red_array)
print(red_array.shape)
```

```python
# the array to split, the number of equally-sized arrays after split, the axis along to split
red_array, green_array, blue_array = np.split(rgb, 3, axis = 2)
print(red_array)
print(red_array.shape) # same shape as the original array 
```

```python
#Trailing dimensions
red_array_2D = red_array.reshape((3,3))
print(red_array_2D)
print(red_array_2D.shape)
```
